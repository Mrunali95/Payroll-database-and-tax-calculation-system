```python
from tkinter import *
from tkinter import ttk
import tkinter.messagebox
import pymysql

class connectorDB:
    
    def __init__(self, root):
        self.root = root
        titlespace = ' '
        self.root.title(102*titlespace+ "Tax Calculator")
        self.root.geometry("850x700+300+0")
        self.root.resizable(width=False, height=False)
        
        Mainframe = Frame(self.root, bd=10, width=800, height=700, relief= RIDGE, bg= 'cadet blue')
        Mainframe.grid()
        
        TitleFrame = Frame(Mainframe, bd=7, width=800, height=100, padx=22, relief= RIDGE)
        TitleFrame.grid(row=0, column=0)
        TopFrame = Frame(Mainframe, bd=5, width=600, height=600, relief= RIDGE)
        TopFrame.grid(row=1, column=0)
        
        LeftFrame = Frame(TopFrame, bd=5, width=800, height=400, relief= RIDGE, padx=2, bg= 'cadet blue')
        LeftFrame.pack(side=LEFT)
        LeftFrame1 = Frame(LeftFrame, bd=5, width=800, height=180, relief= RIDGE, padx=2, pady=14)
        LeftFrame1.pack(side=TOP)
        
        RightFrame = Frame(TopFrame, bd=7, width=90, height=400, relief= RIDGE, padx=2, bg= 'cadet blue')
        RightFrame.pack(side=RIGHT)
        RightFrame1 = Frame(RightFrame, bd=5, width=80, height=300, relief= RIDGE, padx=15, pady=5)
        RightFrame1.pack(side=TOP)
        
        #==================================================================================================
        Agency_Name =StringVar()
        First_Name =StringVar()
        Last_Name =StringVar()
        Agency_Start_Date =StringVar()
        Work_Location_Borough =StringVar()
        Title_Description =StringVar()
        Base_Salary =StringVar() 
        Employee_ID =StringVar() 
        tax =StringVar()
        stax =StringVar()
        
        #==================================================================================================
            
        def iexit():
            iexit = tkinter.messagebox.askyesno("Tax Calculator", "Confirm you want to exit?")
            if iexit >0:
                root.destroy()
                return
            
        def reset():
            self.entAgency_Name.delete(0, END)
            self.entFirst_Name.delete(0, END)
            self.entLast_Name.delete(0, END)
            self.entAgency_Start_Date.delete(0, END)
            self.entWork_Location_Borough.delete(0, END)
            self.entTitle_Description.delete(0, END)
            self.entBase_Salary.delete(0, END)
            self.entEmployee_ID.delete(0, END)
            
        def adddata():
            if First_Name.get() == "" or Last_Name.get() == "" or Base_Salary.get() == "":
                tkinter.messagebox.showerror("MySQL Connection","Enter valid data")
            else:
                sqlcon = pymysql.connect(host='localhost', user='root', password='Mrun@li1995', database='tax')
                cur = sqlcon.cursor()
                cur.execute("Insert into payroll_16 values(%s,%s,%s,%s,%s,%s,%s,%s)",(
                    Agency_Name.get(),
                    First_Name.get(),
                    Last_Name.get(),
                    Agency_Start_Date.get(),
                    Work_Location_Borough.get(),
                    Title_Description.get(),
                    Base_Salary.get(),
                    Employee_ID.get()))
                
                sqlcon.commit()
                sqlcon.close()
                tkinter.messagebox.showinfo("Data Entry Form","Record entered successfully")
                
        def displaydata():
            sqlcon = pymysql.connect(host='localhost', user='root', password='Mrun@li1995', database='tax')
            cur = sqlcon.cursor()
            cur.execute("Select * From payroll_16")
            result = cur.fetchall()
            if len(result) != 0:
                self.payroll_record.delete(*self.payroll_record.get_children()) 
                for row in result:
                    self.payroll_record.insert('',END,values=row)
                
            sqlcon.commit()
            sqlcon.close()
            
        def TraineeInfo(ev):
            viewInfo = self.payroll_record.focus()
            learnerData = self.payroll_record.item(viewInfo)
            row = learnerData['values']
            Agency_Name.set(row[0])
            First_Name.set(row[1])
            Last_Name.set(row[2])
            Agency_Start_Date.set(row[3])
            Work_Location_Borough.set(row[4])
            Title_Description.set(row[5])
            Base_Salary.set(row[6])
            Employee_ID.set(row[7])
            
        def update():
            sqlcon = pymysql.connect(host='localhost', user='root', password='Mrun@li1995', database='tax')
            cur = sqlcon.cursor()
            cur.execute("update payroll_16 set Agency_Name=%s, First_Name=%s, Last_Name=%s, Agency_Start_Date=%s, Work_Location_Borough=%s, Title_Description=%s, Base_Salary=%s where Employee_ID=%s",(
                    Agency_Name.get(),
                    First_Name.get(),
                    Last_Name.get(),
                    Agency_Start_Date.get(),
                    Work_Location_Borough.get(),
                    Title_Description.get(),
                    Base_Salary.get(),
                    Employee_ID.get()))
                
            sqlcon.commit()
            sqlcon.close()
            displaydata()
            tkinter.messagebox.showinfo("Data Update Form","Record updated successfully")
            
        def delete():
            sqlcon = pymysql.connect(host='localhost', user='root', password='Mrun@li1995', database='tax')
            cur = sqlcon.cursor()
            cur.execute("delete from payroll_16 where Employee_ID=%s", Employee_ID.get())
                
            sqlcon.commit()
            sqlcon.close()
            reset()
            displaydata()
            tkinter.messagebox.showinfo("Data delete Form","Record deleted successfully")
            
        def search():
            try:
                sqlcon = pymysql.connect(host='localhost', user='root', password='Mrun@li1995', database='tax')
                cur = sqlcon.cursor()
                cur.execute("select * from payroll_16 where Employee_ID=%s", Employee_ID.get())

                row = cur.fetchone()
                
                Agency_Name.set(row[0])
                First_Name.set(row[1])
                Last_Name.set(row[2])
                Agency_Start_Date.set(row[3])
                Work_Location_Borough.set(row[4])
                Title_Description.set(row[5])
                Base_Salary.set(row[6])
                Employee_ID.set(row[7])

                sqlcon.commit()
            except:
                
                tkinter.messagebox.showinfo("Data Search Form","No such record found")
                reset()
            sqlcon.close()
            
        def caltax():
            sqlcon = pymysql.connect(host='localhost', user='root', password='Mrun@li1995', database='tax')
            cur = sqlcon.cursor()
            cur.execute("select * from payroll_16 where Employee_ID=%s", Employee_ID.get())
            for x in cur:
                bs = (x[6])
                
            bs = float(bs[1:])
                
            t = bs * 22.2/100
            salaryaftertax = bs - t
            
            tax.set(t)
            stax.set(salaryaftertax)
            
            sqlcon.commit()
            sqlcon.close()
            
        #==================================================================================================
        
        self.lbltitle = Label(TitleFrame, font=('Ariel',20, 'bold'), text="Payroll Database and Tax Calculator ", bd=7)
        self.lbltitle.grid(row=0, column=0, padx=132)
        
        self.lblAgency_Name = Label(LeftFrame1, font=('Ariel',10, 'bold'), text="Agency_Name", bd=7)
        self.lblAgency_Name.grid(row=0, column=0, padx=5, sticky=W)
        self.entAgency_Name = Entry(LeftFrame1, font=('Ariel',10, 'bold'), bd=7, justify="left", textvariable=Agency_Name)
        self.entAgency_Name.grid(row=0, column=1, padx=5, sticky=W)
        
        self.lblLast_Name = Label(LeftFrame1, font=('Ariel',10, 'bold'), text="Last Name", bd=7)
        self.lblLast_Name.grid(row=1, column=0, padx=5, sticky=W)
        self.entLast_Name = Entry(LeftFrame1, font=('Ariel',10, 'bold'), bd=7, justify="left", textvariable=Last_Name)
        self.entLast_Name.grid(row=1, column=1, padx=5, sticky=W)
        
        self.lblFirst_Name = Label(LeftFrame1, font=('Ariel',10, 'bold'), text="First Name", bd=7)
        self.lblFirst_Name.grid(row=2, column=0, padx=5, sticky=W)
        self.entFirst_Name = Entry(LeftFrame1, font=('Ariel',10, 'bold'), bd=7, justify="left",textvariable=First_Name)
        self.entFirst_Name.grid(row=2, column=1, padx=5, sticky=W)
        
        self.lblAgency_Start_Date = Label(LeftFrame1, font=('Ariel',10, 'bold'), text="Start Date", bd=7)
        self.lblAgency_Start_Date.grid(row=3, column=0, padx=5, sticky=W)
        self.entAgency_Start_Date = Entry(LeftFrame1, font=('Ariel',10, 'bold'), bd=7, justify="left", textvariable=Agency_Start_Date)
        self.entAgency_Start_Date.grid(row=3, column=1, padx=5, sticky=W)
        
        self.lblWork_Location_Borough = Label(LeftFrame1, font=('Ariel',10, 'bold'), text="Work_Location_Borough", bd=7)
        self.lblWork_Location_Borough.grid(row=4, column=0, padx=5, sticky=W)
        self.entWork_Location_Borough = Entry(LeftFrame1, font=('Ariel',10, 'bold'), bd=7, justify="left", textvariable=Work_Location_Borough)
        self.entWork_Location_Borough.grid(row=4, column=1, padx=5, sticky=W)
        
        self.lblTitle_Description = Label(LeftFrame1, font=('Ariel',10, 'bold'), text="Job Description", bd=7)
        self.lblTitle_Description.grid(row=5, column=0, padx=5, sticky=W)
        self.entTitle_Description = Entry(LeftFrame1, font=('Ariel',10, 'bold'), bd=7, justify="left", textvariable=Title_Description)
        self.entTitle_Description.grid(row=5, column=1, padx=5, sticky=W)
        
        self.lblBase_Salary = Label(LeftFrame1, font=('Ariel',10, 'bold'), text="Base Salary", bd=7)
        self.lblBase_Salary.grid(row=6, column=0, padx=5, sticky=W)
        self.entBase_Salary = Entry(LeftFrame1, font=('Ariel',10, 'bold'), bd=7, justify="left", textvariable=Base_Salary)
        self.entBase_Salary.grid(row= 6, column=1, padx=5, sticky=W)
        
        self.lblEmployee_ID = Label(LeftFrame1, font=('Ariel',10, 'bold'), text="Employee_ID", bd=7)
        self.lblEmployee_ID.grid(row=7, column=0, padx=5, sticky=W)
        self.entEmployee_ID = Entry(LeftFrame1, font=('Ariel',10, 'bold'), bd=7, justify="left", textvariable=Employee_ID)
        self.entEmployee_ID.grid(row= 7, column=1, padx=5, sticky=W)
        
        self.lbltax = Label(RightFrame1, font=('Ariel',10, 'bold'), text="Tax", bd=7)
        self.lbltax.grid(row=8, column=0, padx=5, sticky=W)
        self.enttax = Entry(RightFrame1, font=('Ariel',10, 'bold'), bd=7, justify="left", textvariable=tax)
        self.enttax.grid(row= 9, column=0, padx=5, sticky=W)
        
        self.lblstax = Label(RightFrame1, font=('Ariel',10, 'bold'), text="Salary After Tax", bd=7)
        self.lblstax.grid(row=10, column=0, padx=5, sticky=W)
        self.entstax = Entry(RightFrame1, font=('Ariel',10, 'bold'), bd=7, justify="left", textvariable=stax)
        self.entstax.grid(row= 11, column=0, padx=5, sticky=W)
        
        #===============================================table tree===================================================
        scroll_y = Scrollbar(LeftFrame, orient=VERTICAL)
        self.payroll_record = ttk.Treeview(LeftFrame, height=14, column=('Agency_Name','Last_Name','First_Name', 'Agency_Start_Date',
                                                'Work_Location_Borough', 'Title_Description','Base_Salary','Employee_ID'),yscrollcommand= scroll_y.set)
        scroll_y.pack(side= RIGHT, fill= Y)
        
        self.payroll_record.heading('Agency_Name',text='Agency_Name')
        self.payroll_record.heading('Last_Name',text='Last Name')
        self.payroll_record.heading('First_Name',text='First Name')
        self.payroll_record.heading('Agency_Start_Date',text='Start Date')
        self.payroll_record.heading('Work_Location_Borough',text='Work_Location_Borough')
        self.payroll_record.heading('Title_Description',text='Job Desc')
        self.payroll_record.heading('Base_Salary',text='Base Salary')
        self.payroll_record.heading('Employee_ID',text='ID')
        
        self.payroll_record['show'] = 'headings'
        
        self.payroll_record.column('Agency_Name',width=70)
        self.payroll_record.column('Last_Name',width=70)
        self.payroll_record.column('First_Name',width=70)
        self.payroll_record.column('Agency_Start_Date',width=70)
        self.payroll_record.column('Work_Location_Borough',width=70)
        self.payroll_record.column('Title_Description',width=70)
        self.payroll_record.column('Base_Salary',width=70)
        self.payroll_record.column('Employee_ID',width=70)
        
        self.payroll_record.pack(fill =BOTH, expand=1)
        self.payroll_record.bind("<ButtonRelease-1>",TraineeInfo)
        #displaydata()
        
        #=======================================================================================
        
        self.btnAddNew = Button(RightFrame1, font=('Ariel',12, 'bold'), text="Add New", bd=4, pady=1, padx=24, 
                               width=8, height=2, command=adddata).grid(row=0, column=0, padx=1)
        
        self.btnDisplay = Button(RightFrame1, font=('Ariel',12, 'bold'), text="Display", bd=4, pady=1, padx=24, 
                               width=8, height=2, command=displaydata).grid(row=1, column=0, padx=1)
        
        self.btnUpdate = Button(RightFrame1, font=('Ariel',12, 'bold'), text="Update", bd=4, pady=1, padx=24, 
                               width=8, height=2, command=update).grid(row=2, column=0, padx=1)
        
        self.btnDelete = Button(RightFrame1, font=('Ariel',12, 'bold'), text="Delete", bd=4, pady=1, padx=24, 
                               width=8, height=2, command=delete).grid(row=3, column=0, padx=1)
        
        self.btnSearch = Button(RightFrame1, font=('Ariel',12, 'bold'), text="Search", bd=4, pady=1, padx=24, 
                               width=8, height=2,command=search).grid(row=4, column=0, padx=1)
        
        self.btnReset = Button(RightFrame1, font=('Ariel',12, 'bold'), text="Reset", bd=4, pady=1, padx=24, 
                               width=8, height=2, command=reset).grid(row=5, column=0, padx=1)
        
        self.btnExit = Button(RightFrame1, font=('Ariel',12, 'bold'), text="Exit", bd=4, pady=1, padx=24, 
                               width=8, height=2, command=iexit).grid(row=6, column=0, padx=1)
        
        self.btnTax = Button(RightFrame1, font=('Ariel',12, 'bold'), text="Calculate Tax", bd=4, pady=1, padx=24, 
                               width=8, height=2, command=caltax).grid(row=7, column=0, padx=1)
        
if __name__ == '__main__':
    root = Tk()
    application = connectorDB(root)
    root.mainloop()
```
